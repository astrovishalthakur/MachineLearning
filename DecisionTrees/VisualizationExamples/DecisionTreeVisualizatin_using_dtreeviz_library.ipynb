```python
clas = tree.DecisionTreeClassifier()
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)
```

# 1. Classification

```python
viz = dtreeviz(clas,
               X_train,
               y_train,
               feature_names=iris.feature_names,
               class_names=["setosa", "versicolor", "virginica"],
               scale=1.5)

viz
```

# 3. Horizontal Decision Tree

```python
clas = tree.DecisionTreeClassifier(max_depth=2)

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)



viz = dtreeviz(clas,
               X_train,
               y_train,
               feature_names=iris.feature_names,
               class_names=["setosa", "versicolor", "virginica"],
               orientation="LR") # left to right
viz
```

# 4. Show Prediction Path

```python
clf = tree.DecisionTreeClassifier()

# X_train and y_train are already defined 
# as iris's data and target respectively

clf.fit(X_train, y_train)

X = iris.data[np.random.randint(0, len(iris.data)), :]

viz = dtreeviz(clf,
               X_train,
               y_train,
               feature_names=iris.feature_names,
               class_names=["setosa", "versicolor", "virginica"],
               X=X)

viz
```

# 10. Univariate Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("auto-mpg.csv")
```

```python
df.head()
```

```python
df.rename(columns={"mpg":"MPG", "cylinders":"CYL", "displacement":"ENG", "weight":"WG"}, inplace=True)
```

```python
df.head(3)
```

We will only use some columns for our purpose.

```python
df_cars = df.loc[:, ["MPG", "CYL", "ENG", "WG"]]
```

```python
df_cars
```

### Now we have required data, we can start working.

```python
X, y = df_cars[["WG"]], df_cars["MPG"]

dt = DecisionTreeRegressor(max_depth=3, criterion="mae")
dt.fit(X, y)

fig = plt.figure()
ax = fig.gca()
rtreeviz_univar(dt, X, y, "WG", "MPG", ax=ax)
plt.show()
```

# 11. 3D Regression

```python
from mpl_toolkits.mplot3d import Axes3D

X = df_cars[["WG", "ENG"]]
y = df_cars["MPG"]

dt = DecisionTreeRegressor(max_depth=3, criterion="mae")
dt.fit(X, y)

figsize = (20, 18)
fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(111, projection="3d")

t = rtreeviz_bivar_3D(dt, 
                      X,y,
                      feature_names=["Vehicle Weight", "Horse Power"],
                      target_name="MPG",
                      fontsize=14,
                      elev=20, # this is the height of view point, on increasing the value, graph is shown
                               # from higher view point.
                      azim=25,
                      dist=8.2,
                      show={"splits", "title"},
                      ax = ax)

plt.show()
```
