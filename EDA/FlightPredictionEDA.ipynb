```python
def durationconvert(data):
  min = 0
  lis = data.split(" ")
  if "h" in lis[0]:
    min = int(lis[0].replace("h", "")) * 60
  if "m" in lis[-1]:
    min += int(lis[-1].replace("m", ""))
  return min
```

```python
durationconvert("2h 50m")
```

```python
df["Duration"] = df["Duration"].apply(lambda x: durationconvert(x))
```
