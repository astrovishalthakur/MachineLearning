### We can see that date and time are string by default, we need to covnvert it to datetime object

```python
date.date = pd.to_datetime(date.date)
```

```python
date.info()
```

# Extract year

```python
date["date_year"] = date.date.dt.year
date.sample(5)
```

# Extract month


```python
date["date_month"] = date.date.dt.month
date.sample(5)
```

```python
date["date_month_name"] = date.date.dt.month_name()
date.sample(5)
```

# Extract Days

```python
date["date_day"] = date.date.dt.day

date.sample(5)
```

```python
# day of week

date["day_of_week"] = date.date.dt.dayofweek
```

```python
date.sample(5)
```

```python
# day name

date["day_of_week_name"] = date.date.dt.day_name()

date.drop(columns=["product_id", "city_id", "orders"]).head()
```

```python
# is weekend?

date["date_is_weekend"] = np.where(date.day_of_week_name.isin(["Sunday", "Saturday"]), 1, 0)
# np.where(condition, x, y) ---> meaning if condition is true then return x else return y
date.drop(columns=["product_id", "city_id", "orders"]).head()
```

# Extract week of year

```python
date["week_of_year"] = date.date.dt.week

date.drop(columns=["product_id", "city_id", "orders"]).head()
```

# Extract Quater

```python
date["quater"] = date.date.dt.quarter

date.drop(columns=["product_id", "city_id", "orders"]).head()
```

# Extract Semester

```python
date["semester"] = np.where(date.quater.isin([1, 2]), 1, 2)
date.drop(columns=["product_id", "city_id", "orders"]).head()
```
