```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

import seaborn as sns
from sklearn.impute import SimpleImputer, MissingIndicator
```

```python
url = "https://raw.githubusercontent.com/astrovishalthakur/100-days-of-machine-learning/main/day38-missing-indicator/train.csv"
```

```python
df = pd.read_csv(url, usecols=["Age", "Fare", "Survived"])
```

```python
df.isna().mean()*100
```

```python
df.head()
```

```python
X = df.drop(columns="Survived")
y = df.Survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.head()
```

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf,y_train)

y_pred = clf.predict(X_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
mi = MissingIndicator()

mi.fit(X_train)
```

# This object knows which columns in data have missing values

```python
mi.features_
```

It tells that 0th column, i.e. age has missing values.

```python
X_train_missing = mi.transform(X_train)
```

```python
X_train_missing
```

```python
X_test_missing = mi.transform(X_test)
```

```python
X_train['Age_NA'] = X_train_missing
```

```python
X_test['Age_NA'] = X_test_missing
```

```python
X_train
```

```python
si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# This technique became famous that it is available in simple imputer by name of parameter add_indicator

```python
si = SimpleImputer(add_indicator=True)
```

```python
X_train = si.fit_transform(X_train)
```

```python
X_test = si.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
