```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
url = "https://raw.githubusercontent.com/astrovishalthakur/100-days-of-machine-learning/main/day45-feature-construction-and-feature-splitting/train.csv"
```

```python
df = pd.read_csv(url)[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

# Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1 # adding one to count the passanger as well.
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```
